```python
# importing pandas and dataset file


import pandas as pd

df = pd.read_csv('/content/train.csv')

df.head()
```

a) Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them. Justify your action for this task.


```python
# check what columns have missing values and count of them.


print(df.isnull().sum())

print(df.dtypes)
```

b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

```python
df['Mileage'] = df['Mileage'].str.replace(r'\s+[a-zA-Z/]+', '', regex=True)
df['Engine'] = df['Engine'].str.replace(r'\s+CC', '', regex=True)
df['Power'] = df['Power'].str.replace(r'\s+bhp', '', regex=True)
df['New_Price'] = df['New_Price'].str.replace(r'\s+Lakh', '', regex=True)

df.head()
```

```python
# replace the missing values in Mileage, Engine, Power and Seats columns


# Convert columns to numeric (stripping out non-numeric characters where necessary)
df['Mileage'] = pd.to_numeric(df['Mileage'].str.extract('(\d+\.?\d*)')[0], errors='coerce')
df['Engine'] = pd.to_numeric(df['Engine'].str.extract('(\d+\.?\d*)')[0], errors='coerce')
df['Power'] = pd.to_numeric(df['Power'].str.extract('(\d+\.?\d*)')[0], errors='coerce')

# Fill missing numeric values with median
df['Mileage'].fillna(df['Mileage'].median(), inplace=True)
df['Engine'].fillna(df['Engine'].median(), inplace=True)
df['Power'].fillna(df['Power'].median(), inplace=True)

# Fill missing categorical values with mode
df['Seats'].fillna(df['Seats'].mode()[0], inplace=True)
```

```python
# dropping the 'New_Price' column since there are lot of missing values


df.drop(columns=['New_Price'], inplace=True, errors='ignore')

print(df.isnull().sum())
```

c) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python

fuel_mapping = {'Diesel': 0, 'Petrol': 1, 'Electric': 2}
df['Fuel_Type'] = df['Fuel_Type'].replace(fuel_mapping)

transmission_mapping = {'Manual': 0, 'Automatic': 1}
df['Transmission'] = df['Transmission'].replace(transmission_mapping)

df.head()
```

```python

# getting current year to calculate age of the car


from datetime import datetime

current_year = datetime.now().year
```

d) Create one more feature and add this column to the dataset (you can use mutate function in R for this). For example, you can calculate the current age of the car by subtracting “Year” value from the current year.

```python
df['Age_of_Car'] = current_year - df['Year']


df['Power_per_Liter'] = (df['Power'].astype(float) / df['Mileage'].astype(float)).round(2)


df['Mileage_per_Seat'] = (df['Mileage'].astype(float) / df['Seats'].astype(float)).round(2)


df.head()
```

```python
preprocessed_csv_file = '/content/preprocessed_data.csv'
df.to_csv(preprocessed_csv_file, index=False)
```

e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.

```python
# renaming column


renamed_df = df.rename(columns={'Power_per_Liter': 'Power_per_Fuel'})
renamed_df.head()
```

```python
#Arrange
# sorting the df based on age of the car and Mileage features


sorted_df = df.sort_values(by=['Age_of_Car', 'Mileage'], ascending=[True, False])
sorted_df.head()
```

```python
arranged_csv_file = '/content/arranged_data.csv'
sorted_df.to_csv(arranged_csv_file, index=False)


df['Mileage'] = df['Mileage'].astype(float)
```

```python
#Transform function

# To know the mean mileage based on location


df['Avg_Mileage_Location'] = df.groupby('Location')['Mileage'].transform('mean')
df.head()
```

```python
transformed_csv_file = '/content/transformed_data.csv'
df.to_csv(transformed_csv_file, index=False)
```

```python
#Summarize

# Getting the mean mileage based on fuel type


grouped_df = df.groupby('Fuel_Type').agg({'Mileage': 'mean'}).sort_values(by='Mileage', ascending=False)
grouped_df.head()
```

```python

summary_csv_file = '/content/summary_data.csv'
grouped_df.to_csv(summary_csv_file, index=False)
```
